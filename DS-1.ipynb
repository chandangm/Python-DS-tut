# Stat basics

What can we learn from a group of numbers? - That's the goal of stat. 

Let's start with:
- Mean 
- Median
- Mode

## MEAN

Mean is also called average value. 

Steps:
- Add all the values. 
- Divide by total number of values in it.

eg:
```python
#list holding readings of speed    
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
    
#step1: add all the values:
(99+86+87+88+111+86+103+87+94+78+77+85+86)

#step2: divide by total number of values in it. 
# 1167 - total sum 
# 13 - total numbers in list. 

mean = 1167/13
```

## Median

is the value in the middle, after you have sorted all the values. 

Steps:
- Sort the given list of numbers. 
- find the middle value in it (location/index wise)
- if you have even length list, you will have 2 values as middle, in that case: sum those two and divide (take **mean** on it)

eg:
```python

#take speed from above, sort it 
# you'll get this -> 77, 78, 85, 86, 86, 86, 87, 87, 94, 98, 99, 103

speed.sort()
print(speed)

#find the middle element in it.
#13 elements -> 6 on one side and 6 on another, middle one is 87(index:6, position:7)

```
